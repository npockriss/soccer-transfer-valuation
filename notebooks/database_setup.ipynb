```python
import pandas as pd
import sqlite3

# Load existing CSV
df = pd.read_csv('/Users/noahpockriss/Winter Sprint 2026/data/final_data.csv')

# Create SQLite database connection
conn = sqlite3.connect('final_data.db')

# Write DataFrame to SQLite database
df.to_sql('players', conn, if_exists='replace', index=False)

# Prove it works by running back SQL query
query_result = pd.read_sql_query('SELECT * FROM players LIMIT 5;', conn)

# Close the connection
conn.close()
```
