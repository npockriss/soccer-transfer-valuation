```python
import pandas as pd
import sys
!{sys.executable} -m pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt

#Load clean data file
df = pd.read_csv('/Users/noahpockriss/Winter Sprint 2026/data/final_data.csv')

#Set a nice theme
sns.set_theme(style="whitegrid")
```

```python
plt.figure(figsize=(10,6))

# Filter for players with value < 100 mill euros to keep the chart readable and remove extreme outliers
sns.scatterplot(data=df[df['market_value_in_eur'] < 100_000_000],x='goals', y='market_value_in_eur', hue='age', palette='viridis', alpha=0.7)

plt.title('Do Goals = Value? (Player Valuation vs. Goals Scored)')
plt.xlabel('Goals Scored in Season')
plt.ylabel('Market Value (EUR)')
plt.show()
```

```python
# Select only numeric columns for correlation analysis
features = ['age', 'height_in_cm', 'goals', 'assists', 'minutes_played', 'market_value_in_eur']
corr = df[features].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('What Features Drive Market Value?')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['market_value_in_eur'], bins=50, kde=True, color='green')
plt.title('Distribution of Player Market Values')
plt.xlabel('Market Value (EUR)')
plt.show()
```
